```r
rm(list=ls())
#
#setwd('/Users/zhengwei/Desktop/OneDrive\ -\ Texas\ A&M\ University-Corpus\ Christi/Research/Huiyan_Sang/DecisionTree/MyPackage/sim1_lin/res')
library(parallel)
B=100
idvec<-splitIndices(B, 2)

# Read all 10 files into a list
res_list <- lapply(c(1:8,10), function(i) {
  readRDS(file = paste0("linear_case_part", i, ".rds"))
})

# Function to combine each result slot across all files
combine_results <- function(slot_index) {
  do.call(rbind, lapply(res_list, function(x) x[[slot_index]]))
}

# Function to combine and remove outliers
combine_results <- function(slot_index, method = c("sd", "trim"), trim_prop = 0.05, sd_cut = 3) {
  method <- match.arg(method)
  mat <- do.call(rbind, lapply(res_list, function(x) x[[slot_index]]))

  cleaned <- apply(mat, 2, function(col) {
    if (method == "sd") {
      m <- mean(col, na.rm = TRUE)
      s <- sd(col, na.rm = TRUE)
      outliers <- which(col > (m + sd_cut * s) | col < (m - sd_cut * s))
      if (length(outliers) > 0) {
        cat("Slot", slot_index, "- Deleted values:", col[outliers], "\n")
      }
      col[outliers] <- NA
      mean(col, na.rm = TRUE)
    } else if (method == "trim") {
      mean(col, trim = trim_prop, na.rm = TRUE)
    }
  })

  return(cleaned)
}

                        

######Final table
# Build the final matrix
# Build final matrix with outlier removal method = "trim", trim_prop = 0
mat <- matrix(
  c(
    round(combine_results(1, method = "trim", trim_prop = 0), 4),
    round(combine_results(2, method = "trim", trim_prop = 0), 4),
    round(combine_results(3, method = "trim", trim_prop = 0), 4),
    round(combine_results(4, method = "trim", trim_prop = 0), 4),
    round(combine_results(5, method = "trim", trim_prop = 0), 4)
  ),
  nrow = 5, byrow = TRUE
)



colnames(mat)<-c("Bayes Linear","SEM-SFM","BART-SFM","BART-SFM $(\\beta_0)$")
rownames(mat)<-c("RMSE","BIAS","${\\rm Bias}_v$","${\\rm Bias}_u$", "${\\rm Bias}_{TE}$")

library(xtable)

print(xtable(mat,digits = 4), hline.after=5,sanitize.text.function=function(x){x})

```
